```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
model_checkpoint = 'distilbert-base-uncased-finetuned-sst-2-english'
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(model_checkpoint)
print(f"Tokenizer loaded for model: {model_checkpoint}")
print(f"Model loaded for sequence classification: {model_checkpoint}")
text_input = "This movie is absolutely fantastic! I loved every moment of it."
inputs = tokenizer(text_input, return_tensors='pt', truncation=True)
print("Original text input:")
print(text_input)
print("\nTokenized input (PyTorch tensors):")
print(inputs)
import torch.nn.functional as F
import torch
outputs = model(**inputs)
logits = outputs.logits
probabilities = F.softmax(logits, dim=-1)
predicted_class_id = torch.argmax(probabilities, dim=-1)
print("Model Outputs (logits):")
print(logits)
print("\nProbabilities:")
print(probabilities)
print("\nPredicted Class ID:")
print(predicted_class_id)
predicted_label = model.config.id2label[predicted_class_id.item()]
print(f"Predicted Label: {predicted_label}")
print(f"The sentiment of the input text is: {predicted_label}")
```
